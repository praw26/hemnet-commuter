```python
#center in lat-long, range in metres
center=[45.897158, 1.039057]
#320,400,450
range="150000"

iso_response = ls.calculate_isoline(
                    origin=center,
                    range=range,
                    range_type=RANGE_TYPE.distance,
                    transport_mode=ISOLINE_ROUTING_TRANSPORT_MODE.truck,
                    shape_max_points = 30000
                    )

data = iso_response.to_geojson()

#colour is added to the geoJSON layer like this, to add colour to any other layer, use inspect.colour 
#geo_layer = GeoJSON(data=data, style={"fillColor": "rgba(245, 176, 65, 0.5)", "strokeColor": "black", "opacity": 0.7},
#    show_bubble=True)
geo_layer = GeoJSON(data=data, style={"fillColor": "rgba(11,156,49,0.3)"}, name="Layer-inner")
```
